## _Exercise 2.5_
_Loop through the whole 'output.csv' file and do the daily range calculation. There is a little twist, see if you can figure it out._

```python
# Solution

fid = open('output.csv')
fid.readline()
for i in fid:
    line = i.strip().split(',')
    hi = float(line[2])
    lo = float(line[3])
    print(hi-lo)
```
