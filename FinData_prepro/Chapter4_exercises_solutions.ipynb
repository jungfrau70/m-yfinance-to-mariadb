## Ex 4.2

_Calculate the monthly return and volatility for the stock Tesla (TSLA), and plot them. For the volatility, apply a rolling window of 20._

```python
# Solution

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mydata =  yf.download('TSLA')
```

```python
# Monthly return
ret_m = mydata['Adj Close'].resample('M').ffill().pct_change()

ret_m.plot()
plt.title('Monthly return for Tesla')
plt.grid()
plt.show()
```

```python
# Monthly volatility
vol_m = ret_m.rolling(20).std()
vol_m.plot()
plt.grid()
plt.show()
```

## Ex 4.3

Calculate the correlation between the returns of the gold price (ticker 'GLD'), gold stocks (tickers 'NEM', 'FNV'), and gold ETF (ticker 'GDX').

```python
# Solution

# gold_and_miners = pandas_datareader.get_data_yahoo(['GLD', 'NEM', 'FNV', 'GDX'])
gold_and_miners = yf.download(['GLD', 'NEM', 'FNV', 'GDX'], start='2013-01-01', end='2016-01-27')
ret = gold_and_miners['Adj Close'].pct_change().dropna()
corr = ret.corr()
corr
```
